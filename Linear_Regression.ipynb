##Problema de negócio: 

###Prever o preço das casas com base na renda média da área

```python
#importando as bibliotecas utilizadas
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
#importando os dados
df_houses = pd.read_csv('/content/USA_Housing.csv')
```

###Informações do Dataset


```python
df_houses.info()
```

```python
df_houses.head()
```

###Analise Exploratória

```python
df_houses.describe()
```

```python
sns.pairplot(df_houses);
```

####Analisando os graficos acima percebemos uma correlação positiva entre Avg Area Income e Price, então utilizarei a variável Avg Area Income para predizer o Price.

```python
#Separação da variável independente e dependente
#Por convenção utilizamos letras em caixa alta para representar variáveis independentes e caixa baixa para o target
X = df_houses['Avg. Area Income'].values.reshape(-1,1)
y = df_houses['Price'].values.reshape(-1,1)
```

```python
#Separação das amostras de treino e teste
# 70% treino e 30% teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
#Instaciação e treinamento do modelo
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
#Aplicando o modelo nos dados de test
y_predict = model.predict(X_test)
```

```python
#Avaliando a acurácia do modelo
print(f'Acurácia: {model.score(X_test,y_test)*100:.2f} %')
```

###Após aplicação do modelo percebemos que a acurácia do modelo está bem baixa, apesar de termos uma correlação positiva a disperção dos dados está causando ruídos.

#Analise Exploratória dos dados

```python
#Verificando a distribuição
df_houses['Avg. Area Income'].plot(kind='hist');
```

####Na analise do histograma vemos que os dados estão distribuidos normalmente, entretanto no boxplot percebe-se que há muitos dados que são outliers.
####Irei retirar os outliers para verificar o quanto muda a acurácia do modelo

####Para isso calcularei os valores dos limites inferior (Q1 - 1,5 * ALQ)  e superior (Q3 + 1,5 * ALQ) onde ALQ é a amplitude interquartis dada por Q3 - Q1.

```python
#Pelo describe vemos que Q1(25%) = 61480.562388 e Q3(75%) = 75783.33866
Q1 = 61480.562388
Q3 = 75783.33866
Amp = Q3 - Q1
inf_limit = Q1 - 1.5*Amp
sup_limit = Q3 + 1.5*Amp
```

```python
#Retirando os outliers
df_houses_filt = df_houses[(df_houses['Avg. Area Income']>=inf_limit) & (df_houses['Avg. Area Income']<=sup_limit)]
```

```python
#Separação da variável independente e dependente
X = df_houses_filt['Avg. Area Income'].values.reshape(-1,1)
y = df_houses_filt['Price'].values.reshape(-1,1)
```

```python
#Separação das amostras de treino e teste
# 70% treino e 30% teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
#Instaciação e treinamento do modelo
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
#Aplicando o modelo nos dados de test
y_predict = model.predict(X_test)
```

```python
#Avaliando a acurácia do modelo
print(f'Acurácia: {model.score(X_test,y_test)*100:.2f} %')
```

####Desde modo percebemos que retirar os outliers não causou melhora, pelo contrário, tivemos uma piora significativa no modelo.
####Não há possibilidade de fazer transformações logaritmicas ou polinomiais, pois os dados não apresentam essas tedências

####Dessa forma, concluimos que o valor de acurácia de 42.47% é o melhor valor que podemos chegar neste caso.
